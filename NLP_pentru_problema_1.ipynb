```python
# setul 1 de restrictii
from pyscipopt import Model
from pyscipopt import Model, log, exp

model = Model(problemName="Problema1") # numele modelului (Problema1)

x = model.addVar("t1", lb=-0.4, ub=2, vtype="C") # continua intre-0.4 si 2
y = model.addVar("r2", lb=3.2, ub=10, vtype="C") # continua intre 3.2 si 10
obj = model.addVar("obj", lb=None, vtype="C") # variabila auxiliara pentru obiectiv

# nu putem folosi operatorul clasic pyhton ** pentru ridicarea la putere
# ne vom folosi de relatia 2^x = e^xln2 pentru a face o schimbare de baza
# functia exp(x) calculeaza valoarea lui e la puterea x
# ln(2^x + 2^y) devine ln(exp(x * ln2) + exp(y * ln2))
ln2 = 0.69314718056

model.addCons(obj >= log(exp(x * ln2) + exp(y * ln2))) # adaugam o restrictie neliniara
model.setObjective(obj, "minimize") # minimizarea variabilei auxiliare care este liniara

model.addCons(x + y >= 4.5, name="prag_minim") # restrica sumei temperaturilor

model.optimize()

status = model.getStatus()
print("Status:", status)

if status == "optimal":
    x_val = model.getVal(x)
    y_val =  model.getVal(y)
    print("x =", x_val)
    print("y =", y_val)

    import math
    rezultat = math.log(2**x_val + 2**y_val)
    
    print("ln(2^x + 2^y) = ", rezultat)
else:
    print("Nu s-a gasit o solutie optima.")
```
