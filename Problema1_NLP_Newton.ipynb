```python
# rularea 1
import numpy as np
import random
import time

# definirea functiei si a derivatelor 
# facem schimbare de baza la logartim
ln2 = np.log(2)

def f(t):
    return np.log(2**t[0] + 2**t[1])

# prima derivata
def gradient(t):
    t1, t2 = t
    num = 2**t1 + 2**t2
    g1 = (2**t1 * ln2) / num
    g2 = (2**t2 * ln2) / num
    return np.array([g1, g2])

# derivatele partiale de ordin 2
def hessian(t):
    t1, t2 = t
    num = 2**t1 + 2**t2
    h11 = (ln2**2 * 2**t1 * 2**t2) / (num**2)
    h12 = -(ln2**2 * 2**t1 * 2**t2) / (num**2)
    return np.array([[h11, h12], [h12, h11]])

# verificam daca respecta conditiile
def este_valid(t):
    return -0.4 <= t[0] <= 2 and 3.2 <= t[1] <= 10 and (t[0] + t[1]) >= 4.5

# generam o solutie valida
def genereaza_valid():
    while True:
        t = np.array([random.uniform(-0.4, 2), random.uniform(3.2, 10)])
        if t[0] + t[1] >= 4.5: return t

timp_alocat = 2.0 
start_time = time.time()

x_best = genereaza_valid() # generam prima solutie si o memoram in f_best (solutia optima globala)
f_best = f(x_best)

while time.time() - start_time < timp_alocat:
    x = genereaza_valid()
    
    for _ in range(100): # limita de pasi pentru evitarea buclelor infinite
        g = gradient(x)
        h = hessian(x)
        
        # pasul newton: delta = - H^-1 * g
        try:
            alpha = - np.linalg.solve(h + np.eye(2) * 1e-6, g)
        except np.linalg.LinAlgError:
            break
            
        x_nou = x + alpha
        
        # verificam daca noua pozitia este valida si este mai buna
        if este_valid(x_nou) and f(x_nou) < f(x):
            x = x_nou
        else:
            break
            
    # daca f(x) < f(x*) atunci x* = x
    if f(x) < f_best:
        x_best = x
        f_best = f(x)

print(f"t1 = {x_best[0]:.4f}, t2 = {x_best[1]:.4f}")
print(f"Emisii minime = {f_best:.4f}")
```
