```python
# rularea 1

import random
import math

# functia obiectiv
def objective_function(t1, t2):
    # ln(2^t1 + 2^t2)
    return math.log(2**t1 + 2**t2)

# restrictiile
def is_feasible(t1, t2):
    c1 = -0.4 <= t1 <= 2      
    c2 = 3.2 <= t2 <= 10      
    c3 = (t1 + t2) >= 4.5    
    return c1 and c2 and c3

# parametrii Simulated Annealing
temp = 1.0            # temperatura initiala
cooling_rate = 0.995  # tata de racire
min_temp = 0.0001     # pragul de oprire
step_size = 0.1       # pasul de explorare
iterations = 0

# solutia initiala (aceasta trebuie sa fie fezabila)
t1, t2 = 1.0, 4.0 
while not is_feasible(t1, t2):
    t1 = random.uniform(-0.4, 2)
    t2 = random.uniform(3.2, 10)

current_sol = {"t1": t1, "t2": t2}
current_fitness = objective_function(current_sol["t1"], current_sol["t2"])

best_sol = current_sol.copy()
best_fitness = current_fitness

# bucla principala
while temp > min_temp:
    iterations += 1
    
    new_sol = current_sol.copy()
    var_to_change = random.choice(["t1", "t2"])
    
    new_sol[var_to_change] += random.uniform(-step_size, step_size)
    
    # verificam daca noua solutie respecta restrictiile
    if is_feasible(new_sol["t1"], new_sol["t2"]):
        new_fitness = objective_function(new_sol["t1"], new_sol["t2"])
        
        delta_e = new_fitness - current_fitness
        
        if delta_e < 0 or random.random() < math.exp(-delta_e / temp):
            current_sol = new_sol
            current_fitness = new_fitness
            
            if current_fitness < best_fitness:
                best_sol = current_sol.copy()
                best_fitness = current_fitness
    
    temp *= cooling_rate

print(f"--- Rezultat Simulated Annealing ({iterations} iteratii) ---")
print(f"solutie: t1 = {best_sol['t1']:.4f}, t2 = {best_sol['t2']:.4f}")
print(f"nivel minim emisii: {best_fitness:.6f}")
```
